```python
# removing the non-numeric row
df2 = df1_copy.drop(df1_copy.index[9990])
df2
```

```python
df2["Reviews"] = df2["Reviews"].astype(int)
```

```python
df2.info()
```

```python
df2["Size"].unique()
```

```python
# 1 mb= 1024 kb
item = "252M"
item[-1]
```

```python
def size_process(x):
    if x[-1] == "M":
        res = float(str(x).replace("M",""))
        return res* 1024
    elif x[-1] == "k":
        return float(str(x).replace("k",""))
    else:
        return np.NaN
```

```python
df2["Size"] = df2["Size"].apply(size_process)
```

```python
df2.Size = df2.Size.astype(float)
df2.info()
```

```python
df2["Installs"]
```

```python
df2["Installs"] = df2["Installs"].str.replace("+","").str.replace(",","")
```

```python
df2["Installs"] = df2["Installs"].astype(int)
df2.info()
```

```python
df2["Price"].unique()
```

```python
char_to_remove =["+",",","$"]
for char in char_to_remove:
    df2["Price"] = df2["Price"].str.replace(char,"")
```

```python
df2["Price"] = df2["Price"].astype(float)
df2.info()
```

```python
df2["Last Updated"]
```

```python
df2["Last Updated"] = pd.to_datetime(df2["Last Updated"])
df2.info()
```

```python
df2["Last Updated Month"] = df2["Last Updated"].dt.month
df2["Last Updated Year"] = df2["Last Updated"].dt.year
df2["Last Updated day"] = df2["Last Updated"].dt.day
df2.drop("Last Updated",axis =1,inplace = True)
df2.info()
```
